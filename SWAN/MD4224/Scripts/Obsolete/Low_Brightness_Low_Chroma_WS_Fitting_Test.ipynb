## Original first_filter: set all points in window to 0.0

```python
# We have ~10174 points along the x axis from -20mm to +20mm
# Need to remove those points that include the large peak (between -25000 and +25000)
def first_filter(ndf, WS = '65.H', limit_min = -23E3, limit_max = 40E3):
    df = ndf.copy()
    # Array for each shot position and profile
    h_positions=[]
    h_profiles=[]
    
    position_label = 'PR.BWS.'+WS+'_ROT:PROF_POSITION_IN'
    profile_label = 'PR.BWS.'+WS+'_ROT:PROF_DATA_IN'

    # Iterate over dataframe
    for i in range(0, len(df), 1):
        position=[]
        profile=[]

        # Iterate over position data
        for j in range(0, len(df[position_label].iloc()[i]), 1):

            # Conditional to filter initial noise peak
            if (df[position_label].iloc()[i][j] < limit_max) and (df[position_label].iloc()[i][j] > limit_min):

                # Minimum value of profile
                offset_y = min(df[profile_label].iloc()[i])

                # Store data in arrays
                position.append(df[position_label].iloc()[i][j]/1000)
                profile.append(df[profile_label].iloc()[i][j] - offset_y)
            else: # store zeroes
                position.append(df[position_label].iloc()[i][j]/1000)
                profile.append(0.0)

        # Array of arrays for dataframe
        h_positions.append(np.array(position))
        h_profiles.append(np.array(profile))

    # Set values in dataframe
    df['position '+WS[-1]] = h_positions
    df['profile '+WS[-1]] = h_profiles
    
    return df
```

```python
def gaussian_5_parameters(x, c, m, A, mu, sig):
    """gaussian_5_parameter(x, c, m, A, mu, sig)"""
    return c+m*x+A/np.sqrt(2*np.pi)/sig*np.exp(-np.power(x - mu, 2.) / (2 * np.power(sig, 2.)))

def gaussian_4_parameters(x, m, A, mu, sig):
    """gaussian_4_parameter(x, m, A, mu, sig)"""
    return m*x+A/np.sqrt(2*np.pi)/sig*np.exp(-np.power(x - mu, 2.) / (2 * np.power(sig, 2.)))
```

```python
# This function takes a list of absolute paths to each file (use for example glob.glob('path_to_files/*.tfs'))
def create_dictionary_of_tfs_table_names(myTFSFiles):
    # We assume that we only vary tunes in one plane (vertical)
    TFSfiles = {}
    for item in myTFSFiles:
        absolute_filename = item                          # filepath
        filename = item.split('/')[-1]                    # filename
        integer = filename.split('p')[1].split('_')[1]    # integer Qy
        fractional = filename.split('p')[2].split('.')[0] # fractional Qy
        qy_str = (integer + '.' + fractional)             # concatenate strings
        qy_float = float(qy_str)                          # convert to float

        TFSfiles.update({qy_str : absolute_filename})
    return TFSfiles   
```

```python
tds = {}
td_peakdensities = {}
for item in myTomoEvalFiles:
    tomoData = extract_long_eval(item)
    
    tomo_time = tomoData['time'] #myToolbox.unixtime2datetime(tomoData['time'])
    
    tds[tomo_time] = tomoData['bunchingfactor']
    td_peakdensities[tomo_time] = tomoData['peakcurrent'] / (beta * c)
    
tomo_times = np.array(sorted(tds.keys()))
```

```python
tunes = []
bfs = []
peakdensities = []
for index, row in df_new.iterrows():
    tunes.append(6 + round_sig(row['Qy'], 1))
    
    
    tomo_time_id = np.argmin(np.abs(tomo_times - (row['cycleStamp'] - 2*3600)))
    bfs.append(tds[tomo_times[tomo_time_id]])
    peakdensities.append(td_peakdensities[tomo_times[tomo_time_id]])
```

```python
alpha_c = 0.027
R = 100
D_x_est = alpha_c * R # average D_x around the ring
Qx = 6 + df_new['Qx']
Qy = 6 + df_new['Qy']
sig_x_eff = df_new['sig H']*1e-3 * D_x_est * Qy/Qx / (np.sqrt(df_new['betx']) * df_new['dx'])
sig_y_eff = df_new['sig V']*1e-3 / np.sqrt(df_new['bety'])

lambda_max = np.median(peakdensities)
K_sc = e * lambda_max / (2*np.pi*epsilon_0 * beta**2 * gamma**3 * m_p * c**2)
dQy_spread_kv = K_sc * R / (4*sig_y_eff * (sig_x_eff + sig_y_eff))
dQy_spread_gauss = 2 * dQy_spread_kv
```
