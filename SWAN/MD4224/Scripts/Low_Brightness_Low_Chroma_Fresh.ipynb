```python
myTFSFiles=glob.glob('/eos/user/h/harafiqu/SWAN_projects/PS/MD4224_Twiss/*.tfs')
myTomoEvalFiles=glob.glob('/eos/user/h/harafiqu/SWAN_projects/PS/MD4224_2018_10_22_Low_Brightness_Low_Chroma/tomo/*_eval.dat')
#myTFSFiles
#myTomoEvalFiles
```

```python
# This function takes a list of absolute paths to each file (use for example glob.glob('path_to_files/*.tfs'))
def create_dictionary_of_tfs_table_names(myTFSFiles):
    # We assume that we only vary tunes in one plane (vertical)
    TFSfiles = {}
    for item in myTFSFiles:
        absolute_filename = item                          # filepath
        filename = item.split('/')[-1]                    # filename
        integer = filename.split('p')[1].split('_')[1]    # integer Qy
        fractional = filename.split('p')[2].split('.')[0] # fractional Qy
        qy_str = (integer + '.' + fractional)             # concatenate strings
        qy_float = float(qy_str)                          # convert to float

        TFSfiles.update({qy_str : absolute_filename})
    return TFSfiles   
```

```python
TFSfiles = create_dictionary_of_tfs_table_names(myTFSFiles)
TFSfiles
```

# Main function to calculate maximum tunespread 

```python
%run /eos/user/h/harafiqu/SWAN_projects/PS/Tools/Tunespread_Calculator.ipynb
```

```python
df_new = df_from_pickle

df_new["dQx"] = np.nan
df_new["dQy"] = np.nan

beta=0.91444281513833
gamma=2.4708737618826

from math import log10, floor
def round_sig(x, sig=2):
    return round(x, sig-int(floor(log10(abs(x))))-1)

# find first tomo timestamp
first_tomo_timestamp = float(extract_long_eval(myTomoEvalFiles[0])['time'])
print 'first recorded tomo at t = ',  myToolbox.unixtime2datetime(first_tomo_timestamp)

# find first data timestamp
row = next(df_new.iterrows())[1]
first_data_timestamp = row['cf_timeStamp']
print 'first recorded data at t = ', first_data_timestamp
tunespread=[]

# Iterate over each row in the dataframe
for index, row in df_new.iterrows():
    
    integer_tune = 6
    fractional_tune = round_sig(row['Qy'],1)
    tune = integer_tune + fractional_tune
    
    # we want to use this twiss file:
    absolute_tfs_filename = TFSfiles[str(tune)]
    
    # Iterate through tomo files to find corresponding data
    for item in myTomoEvalFiles:
        tomoData = extract_long_eval(item)
        
        # use cyclestamp to find correct tomo file
        #if abs(tomoData['time'] - (row['cycleStamp'])) < 10:
        tomo_time = myToolbox.unixtime2datetime(tomoData['time'])
        if abs( tomo_time - (row['cf_timeStamp'])) < pnd.Timedelta('0 days 00:00:20'):            
            
            print '\ntomo file matched at tomo = ', tomo_time, ', data time = ', row['cf_timeStamp'], ', diff = ', (tomo_time - row['cf_timeStamp'])
            
            # We are at the correct row in the dataframe, and have identified the correct tomo file
            # Let's do some tunespread calculation!
            
            # Read TFS table
            tfsData = read_tfs_table_return_data(absolute_tfs_filename)
            
            # Create dictionary for tunespread tool:
            #make_tst_inputs(intensity, deltap, emit_geo_x, emit_geo_y, std_x_div_by_Dx, lorentz_beta=0.91444281513833, lorentz_gamma=2.4708737618826, mass=m_p):
            # need the sigma of the wirescanner profile in metres
            inputData = make_tst_inputs(row['intensity'], tomoData['deltap'], (row['emittance H']*1E-6/beta*gamma), (row['emittance V']*1E-6/beta*gamma), ((row['sig H']*1E-3)/row['dx'])  )

            # Run tunespread function for this shot
            #calculate_exact_tunespread(data, inputs, beta=0.91444281513833, gamma=2.4708737618826, f_verbose=False):
            tunespread.append(calculate_exact_tunespread(tfsData, inputData, tomoData, f_verbose=True))
            print tunespread[-1]
            df_new.set_value(index, 'dQx', tunespread[-1][0])
            df_new.set_value(index, 'dQy', tunespread[-1][1])
            #df_new.iloc[[index],['dQx']] = tunespread[-1][0]
            #df_new.iloc[[index],['dQy']] = tunespread[-1][1]
            #row['dQx'][index]=(tunespread[-1][0])
            #row['dQy'].replace(tunespread[-1][1])
        else: 
            pass
        

print 'length of matched tomo + df data (tunespreads) = ', len(tunespread)
```
