## Plot Mean Intensities with Error Bars

```python
# Plot averaged data points with error bars
x_data = np.linspace(6040, 6105, 14);
x_data = x_data/1000;

y_data = []
y_error = []

for i in x_data:
    
    upper_lim = i+0.0024999999999 - 6
    lower_lim = i-0.0024999999999 - 6
    
    y_data.append( np.mean( df_second['intensity'][(df_second['Qy']>lower_lim)][(df_second['Qy']<upper_lim)] ) )
    
    y_error.append( np.std( df_second['intensity'][(df_second['Qy']>lower_lim)][(df_second['Qy']<upper_lim)] ) )
    
    
fig=plt.figure(figsize=(6, 3), dpi= 200, facecolor='w', edgecolor='k');

plt.errorbar(x_data, y_data, y_error, linestyle='None', marker='o', capsize=3, color='b')

plt.xlabel('Qx');
plt.ylabel('Intensity [protons]');
plt.title('Low Brightness, Natural Chroma Scan: Intensity')
plt.ylim(1E10,6E11)
#plt.xlim(6.05, 6.25)
#plt.legend()
plt.grid();
fig.savefig('MD4224_LB_NC_Intensity_Errors.png');
```
